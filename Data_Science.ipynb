```python
import csv
import urllib.request
import numpy as np
import matplotlib.pyplot as plt

url='https://data.epa.gov.tw/api/v2/aqx_p_434?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=monitordate%20desc&format=CSV'
webpage=urllib.request.urlopen(url)
data=csv.reader(webpage.read().decode('utf-8').splitlines())

SO2=[]
CO=[]
PM10=[]
NO2=[]
PM25=[]
for i in data:
  if i[1]=="埔里":
    SO2.append(int(i[4]))
    CO.append(int(i[5]))
    PM10.append(int(i[7]))
    NO2.append(int(i[8]))
    PM25.append(int(i[10]))
x=np.arange(1,14,1)
```

```python
plt.figure(figsize=(16,7))
plt.scatter(x,SO2,color='r',label="SO2")
plt.scatter(x,CO,color='y',label="CO")
plt.scatter(x,PM10,color='g',label="PM10")
plt.scatter(x,NO2,color='b',label="NO2")
plt.scatter(x,PM25,color='m',label="PM2.5")
plt.title("Others for one siteid")
plt.xlabel("Day")
plt.ylabel("values")
plt.grid()
plt.legend()
```

```python
url='https://data.epa.gov.tw/api/v2/aqx_p_434?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=monitordate%20desc&format=CSV'#CSV檔的網址
webpage=urllib.request.urlopen(url)  #開啟網頁
data=csv.reader(webpage.read().decode('utf-8').splitlines()) #讀取資料到data陣列中

aqi1=[]
aqi2=[]
aqi3=[]
for i in data:
  if i[1]=="復興":
    aqi1.append(int(i[3]))
  if i[1]=="彰化":
    aqi2.append(int(i[3]))
  if i[1]=="花蓮":
    aqi3.append(int(i[3]))
aqi1.reverse()
aqi2.reverse()
aqi3.reverse()

x=np.arange(1,14,1)
plt.figure(figsize=(20,15))

plt.subplot(3,2,1)
plt.plot(x,aqi1,'r-o',label="Fuxing")
plt.plot(x,aqi2,'g-D',label="Changhua")
plt.plot(x,aqi3,'b-^',label="Hualien")
plt.title("Daily AQI")
plt.xlabel("DAY(111.12)")
plt.ylabel("AQI")
plt.grid()
plt.legend()

plt.subplot(3,2,2)
plt.plot(x,aqi1,'r-o',label="Fuxing")
plt.title("Fuxing Siteid")
plt.xlabel("DAY(111.12)")
plt.ylabel("AQI")
plt.grid()
plt.legend()


plt.subplot(3,2,3)
plt.plot(x,aqi2,'g-D',label="Changhua")
plt.title("Changhua Siteid")
plt.xlabel("DAY(111.12)")
plt.ylabel("AQI")
plt.grid()
plt.legend()

plt.subplot(3,2,4)
plt.plot(x,aqi3,'b-^',label="Hualien")
plt.title("Hualien Siteid")
plt.xlabel("DAY(111.12)")
plt.ylabel("AQI")
plt.grid()
plt.legend()

```
